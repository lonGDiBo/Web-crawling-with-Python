### Scraping  jobs recuitment web site 

```python
import requests  as rq
from bs4 import BeautifulSoup
```

```python
url = 'https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=data+engineering&txtLocation='

page = rq.get(url)

soup = BeautifulSoup(page.text, 'html' )

soup
```

```python
jobs = soup.find_all('li',class_= 'clearfix job-bx wht-shd-bx')

count = 0

print('SKill that you are not familiar whith')
unfamiliar = input('>')
print(f'Filtering out {unfamiliar} ')

# Only take jobs posted a few days ago
for job in jobs:
    job_published_date = job.find('span',class_= 'sim-posted').text.strip()
    
    if 'few' in job_published_date:
        job_company = job.find('h3',class_= 'joblist-comp-name').text.strip()
        job_company = job_company.split('\n')[0].strip()
        skills = job.find('span',class_ = 'srp-skills').text.replace(" ","")
        skills = skills.strip()
        more_info = job.header.h2.a['href']       
        
        # Filler out unfamiliar skills
        if unfamiliar not in skills:
          print(f'''
          Company_name: {job_company}
          Required skills: {skills}
          More_info :{more_info}
            ''')
          print('')
        count = count + 1
    
if count == 0:
  print("There have been no jobs posted in the last few days")
```
